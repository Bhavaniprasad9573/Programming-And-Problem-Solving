```python
def addrowcsv(filename, rowdata):
    df = readCSV(filename) # get data from csv 
    df.loc[len(df)] = rowdata # add a row to the dataframe
    df.to_csv('DataFiles/income.csv', index=False) # send the dataframe back to csv
    return df
rowdata = ['','01010101', 'statement', 524354, 634556, 546356, 54353, 564663, 73536, 353546, 543543, 65743435]
addrowcsv('DataFiles/income.csv', rowdata)   
```

### Problem 2:

Define a function to update/modify a row in csv file

#### Constraints:

#### Test Cases:

```python
def updaterowcsv(filename, rowindex, rowdata):
    df = readCSV(filename)
    df.loc[rowindex] = rowdata
    df.to_csv(filename, index=False)
    return df
rowdata = ['040254965', 'florida', 5354, 634556, 546356, 54353, 564663, 73536, 353546, 543543, 65743435]


updaterowcsv('DataFiles/income.csv', 6, rowdata)
```

### Problem 3:
#### Problem Statement
delete a row from csv file

#### Constraints:

#### Test Cases:

```python
def deleterowcsv(filename, rowindex):
    df = readCSV(filename)
    df = df.drop(rowindex)
    df.to_csv(filename, index=False)
    return df

```

```python
filename = 'DataFiles/income.csv'
rowindex = 4
deleterowcsv(filename, rowindex)
```

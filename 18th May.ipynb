```python
import numpy as np

t = np.arange(0.0, 2.0, 0.1)
v = 1 + np.sin(2 * np.pi * t)
plt.plot(t, v)
plt.grid()
plt.xlabel('time')
plt.ylabel('voltage')
plt.savefig('figures/time-voltage.png')
plt.show
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab

df = pd.read_csv('Python Notebooks/DataFiles/income.csv')
y= df.values[3]
x = df.columns
```

```python
years=[]
for i in range (2,len(x)):
    years.append(x[i])
years
income=[]
for i in range(2,len(y)):
    income.append(y[i])
    
plt.xlabel('years')
plt.ylabel('income')
plt.title('Income of Arizona')
plt.plot(years,income)
plt.grid()
plt.show
```

```python
width=1/2
plt.bar(years,income,width,color='blue')
plt.title('Income of Arizona')
plt.xlabel('years')
plt.ylabel('income')
plt.savefig('figures/income of arizona.png')
```
